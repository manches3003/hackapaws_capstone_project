# Task
Analyze and compare two datasets, "death_records.csv" and "medicine_sales.csv", by loading them, cleaning and standardizing 'Year' and 'Disease' columns, grouping and aggregating by 'Year' and 'Disease', merging the datasets, calculating average recovery rate per disease per year, change in recovery rate before vs. after 2021, and correlation between medicine sales and recovery rate. Finally, visualize the recovery rate trend, compare average medicine units sold vs recovery rate using a bar chart, and show correlations between numeric columns using a heatmap, and print a summary of diseases with the most improved recovery rates after 2021.

## Load the datasets

### Subtask:
Load the `death_records.csv` and `medicine_sales.csv` files into pandas DataFrames.


**Reasoning**:
The first step is to load the two specified CSV files into pandas DataFrames as instructed.



```python
df_death_records = pd.read_csv('/content/us_death_records_by kaushal patidar.csv')
df_medicine_sales = pd.read_csv('/content/unified_health_2020_2025_realnames.csv')

display(df_death_records.head())
display(df_medicine_sales.head())
```

## Clean and standardize data

### Subtask:
Clean and standardize the "Year" and "Disease" columns in both DataFrames to ensure consistency for merging.


**Reasoning**:
Inspect the unique values and data types of the 'Year' columns in both dataframes.



```python
print("Unique values and data type of 'Year' in df_death_records:")
print(df_death_records['Year'].value_counts())
print(df_death_records['Year'].dtype)

print("\nUnique values and data type of 'Year' in df_medicine_sales:")
print(df_medicine_sales['Year'].value_counts())
print(df_medicine_sales['Year'].dtype)
```

**Reasoning**:
Inspect the unique values in the disease-related columns of both dataframes to identify inconsistencies.



```python
print("Unique values in 'Cause_of_Death' in df_death_records:")
print(df_death_records['Cause_of_Death'].unique())

print("\nUnique values in 'Disease_Name' in df_medicine_sales:")
print(df_medicine_sales['Disease_Name'].unique())
```

**Reasoning**:
Standardize the disease names by renaming the 'Cause_of_Death' column and mapping similar disease names to consistent terms.



```python
df_death_records = df_death_records.rename(columns={'Cause_of_Death': 'Disease_Name'})

disease_mapping = {
    'Accidents (Unintentional injuries)': 'Other',
    'Chronic Lower Respiratory Diseases': 'Other',
    'Septicemia / Infections': 'Other',
    'Influenza & Pneumonia': 'Influenza',
    "Alzheimer's disease": "Alzheimer's"
}

df_death_records['Disease_Name'] = df_death_records['Disease_Name'].replace(disease_mapping)

print("Unique values in 'Disease_Name' in df_death_records after standardization:")
print(df_death_records['Disease_Name'].unique())

print("\nUnique values in 'Disease_Name' in df_medicine_sales:")
print(df_medicine_sales['Disease_Name'].unique())
```

## Group and aggregate data

### Subtask:
Group and aggregate each DataFrame by "Year" and "Disease" to prepare for merging.


**Reasoning**:
Group df_death_records by 'Year' and 'Disease_Name' and calculate the sum of 'Death_Count', then reset the index.



```python
df_deaths_aggregated = df_death_records.groupby(['Year', 'Disease_Name'])['Death_Count'].sum().reset_index()
display(df_deaths_aggregated.head())
```

**Reasoning**:
Group df_medicine_sales by 'Year' and 'Disease_Name' and calculate the sum of 'Units_Sold' and the mean of 'Recovery_Rate_by_Disease_Year', then reset the index.



```python
df_medicine_aggregated = df_medicine_sales.groupby(['Year', 'Disease_Name']).agg(
    Total_Units_Sold=('Units_Sold', 'sum'),
    Avg_Recovery_Rate=('Recovery_Rate_by_Disease_Year', 'mean')
).reset_index()
display(df_medicine_aggregated.head())
```

## Merge the datasets

### Subtask:
Merge the aggregated DataFrames based on "Year" and "Disease".


**Reasoning**:
Merge the two aggregated dataframes based on the common columns 'Year' and 'Disease_Name' to combine the death count and medicine sales/recovery rate information.



```python
df_merged_data = pd.merge(df_deaths_aggregated, df_medicine_aggregated, on=['Year', 'Disease_Name'], how='inner')
display(df_merged_data.head())
```

## Calculate key metrics

### Subtask:
Calculate average recovery rate per disease per year, the change in recovery rate before vs. after 2021, and the correlation between medicine sales and recovery rate.


**Reasoning**:
Calculate the average recovery rate per disease per year, the change in recovery rate before vs. after 2021, and the correlation between medicine sales and recovery rate, combining multiple steps into one code block as requested.



```python
# Calculate average recovery rate per disease across all years in the merged data
avg_recovery_rate_per_disease_overall = df_merged_data.groupby('Disease_Name')['Avg_Recovery_Rate'].mean().reset_index()
print("Average Recovery Rate per Disease (Overall):")
display(avg_recovery_rate_per_disease_overall)

# Create dataframes for years before/in 2021 and after 2021
df_pre_2021 = df_merged_data[df_merged_data['Year'] <= 2021]
df_post_2021 = df_merged_data[df_merged_data['Year'] > 2021]

# Calculate average recovery rate per disease for pre-2021
avg_recovery_rate_pre_2021 = df_pre_2021.groupby('Disease_Name')['Avg_Recovery_Rate'].mean().reset_index()
avg_recovery_rate_pre_2021 = avg_recovery_rate_pre_2021.rename(columns={'Avg_Recovery_Rate': 'Avg_Recovery_Rate_Pre_2021'})
print("\nAverage Recovery Rate per Disease (Pre-2021):")
display(avg_recovery_rate_pre_2021)

# Calculate average recovery rate per disease for post-2021
avg_recovery_rate_post_2021 = df_post_2021.groupby('Disease_Name')['Avg_Recovery_Rate'].mean().reset_index()
avg_recovery_rate_post_2021 = avg_recovery_rate_post_2021.rename(columns={'Avg_Recovery_Rate': 'Avg_Recovery_Rate_Post_2021'})
print("\nAverage Recovery Rate per Disease (Post-2021):")
display(avg_recovery_rate_post_2021)

# Calculate the change in average recovery rate (Post-2021 - Pre-2021)
recovery_rate_change = pd.merge(avg_recovery_rate_pre_2021, avg_recovery_rate_post_2021, on='Disease_Name', how='outer')
recovery_rate_change['Recovery_Rate_Change'] = recovery_rate_change['Avg_Recovery_Rate_Post_2021'] - recovery_rate_change['Avg_Recovery_Rate_Pre_2021']
print("\nChange in Average Recovery Rate per Disease (Post-2021 vs Pre-2021):")
display(recovery_rate_change)

# Calculate the Pearson correlation coefficient between 'Total_Units_Sold' and 'Avg_Recovery_Rate'
correlation = df_merged_data['Total_Units_Sold'].corr(df_merged_data['Avg_Recovery_Rate'])
print(f"\nPearson correlation between Total_Units_Sold and Avg_Recovery_Rate: {correlation}")
```

## Analyze impact of government spending

### Subtask:
Investigate the relationship between government spending and recovery rates or medicine sales, potentially by disease or region.

**Reasoning**:
Merge the aggregated data with the original medicine sales data to include the government spending column, calculate correlations between government spending and recovery rate/medicine sales, and then visualize these relationships using scatter plots, combining multiple steps into a single code block.

```python
# 1. Merge df_merged_data with df_medicine_sales to include 'Govt_Spending_Million'
# Aggregate df_medicine_sales by Year and Disease_Name to get Avg_Govt_Spending_Million
df_spending_aggregated = df_medicine_sales.groupby(['Year', 'Disease_Name'])['Govt_Spending_Million'].mean().reset_index()

# Merge with df_merged_data
df_merged_spending = pd.merge(df_merged_data, df_spending_aggregated, on=['Year', 'Disease_Name'], how='left')
print("Merged DataFrame with Govt_Spending_Million:")
display(df_merged_spending.head())

# 2. Calculate the correlation between 'Govt_Spending_Million' and 'Avg_Recovery_Rate'
correlation_spending_recovery = df_merged_spending['Govt_Spending_Million'].corr(df_merged_spending['Avg_Recovery_Rate'])
print(f"\nPearson correlation between Govt_Spending_Million and Avg_Recovery_Rate: {correlation_spending_recovery}")

# 3. Calculate the correlation between 'Govt_Spending_Million' and 'Total_Units_Sold'
correlation_spending_sales = df_merged_spending['Govt_Spending_Million'].corr(df_merged_spending['Total_Units_Sold'])
print(f"Pearson correlation between Govt_Spending_Million and Total_Units_Sold: {correlation_spending_sales}")

# 4. Calculate the correlation between 'Govt_Spending_Million' and 'Avg_Recovery_Rate' for each disease
disease_spending_recovery_correlation_list = []

for disease in df_merged_spending['Disease_Name'].unique():
    df_single_disease = df_merged_spending[df_merged_spending['Disease_Name'] == disease].dropna(subset=['Govt_Spending_Million', 'Avg_Recovery_Rate'])
    if len(df_single_disease) > 1: # Need at least 2 data points with non-null values to calculate correlation
        correlation = df_single_disease['Govt_Spending_Million'].corr(df_single_disease['Avg_Recovery_Rate'])
        disease_spending_recovery_correlation_list.append({'Disease_Name': disease, 'Correlation_Spending_Recovery': correlation})

df_disease_spending_recovery_correlation = pd.DataFrame(disease_spending_recovery_correlation_list)
df_disease_spending_recovery_correlation = df_disease_spending_recovery_correlation.sort_values(by='Correlation_Spending_Recovery', ascending=False)

print("\nCorrelation between Govt_Spending_Million and Avg_Recovery_Rate per Disease:")
display(df_disease_spending_recovery_correlation)

# 5. Scatter plot: Govt_Spending_Million vs Avg_Recovery_Rate
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df_merged_spending, x='Govt_Spending_Million', y='Avg_Recovery_Rate', hue='Disease_Name', s=100)
plt.title('Government Spending vs. Average Recovery Rate per Disease')
plt.xlabel('Government Spending (Million $)')
plt.ylabel('Average Recovery Rate')
plt.legend(title='Disease Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# 6. Scatter plot: Govt_Spending_Million vs Total_Units_Sold
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df_merged_spending, x='Govt_Spending_Million', y='Total_Units_Sold', hue='Disease_Name', s=100)
plt.title('Government Spending vs. Total Units Sold per Disease')
plt.xlabel('Government Spending (Million $)')
plt.ylabel('Total Units Sold')
plt.legend(title='Disease Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Load the datasets
try:
    df_death_records = pd.read_csv('/content/us_death_records_by kaushal patidar.csv')
    df_medicine_sales = pd.read_csv('/content/unified_health_2020_2025_realnames.csv')
except FileNotFoundError as e:
    print(f"Error loading data: {e}. Please ensure the files are uploaded correctly.")
    raise # Re-raise the error to stop execution if files are not found


# Clean and standardize data
df_death_records = df_death_records.rename(columns={'Cause_of_Death': 'Disease_Name'})

disease_mapping = {
    'Accidents (Unintentional injuries)': 'Other',
    'Chronic Lower Respiratory Diseases': 'Other',
    'Septicemia / Infections': 'Other',
    'Influenza & Pneumonia': 'Influenza',
    "Alzheimer's disease": "Alzheimer's"
}

df_death_records['Disease_Name'] = df_death_records['Disease_Name'].replace(disease_mapping)

# Group and aggregate data
df_deaths_aggregated = df_death_records.groupby(['Year', 'Disease_Name'])['Death_Count'].sum().reset_index()

df_medicine_aggregated = df_medicine_sales.groupby(['Year', 'Disease_Name']).agg(
    Total_Units_Sold=('Units_Sold', 'sum'),
    Avg_Recovery_Rate=('Recovery_Rate_by_Disease_Year', 'mean'),
    Avg_Govt_Spending_Million=('Govt_Spending_Million', 'mean') # Include spending for later analysis
).reset_index()

# Merge the datasets
df_merged_data = pd.merge(df_deaths_aggregated, df_medicine_aggregated, on=['Year', 'Disease_Name'], how='inner')

# Calculate correlation matrix for heatmap
numeric_cols = df_merged_data.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df_merged_data[numeric_cols].corr()

# Calculate average recovery rate per disease across all years
avg_recovery_rate_per_disease_overall = df_merged_data.groupby('Disease_Name')['Avg_Recovery_Rate'].mean().reset_index()

# Calculate the change in average recovery rate (Post-2021 - Pre-2021)
df_pre_2021 = df_merged_data[df_merged_data['Year'] <= 2021]
df_post_2021 = df_merged_data[df_merged_data['Year'] > 2021]

avg_recovery_rate_pre_2021 = df_pre_2021.groupby('Disease_Name')['Avg_Recovery_Rate'].mean().reset_index()
avg_recovery_rate_pre_2021 = avg_recovery_rate_pre_2021.rename(columns={'Avg_Recovery_Rate': 'Avg_Recovery_Rate_Pre_2021'})

avg_recovery_rate_post_2021 = df_post_2021.groupby('Disease_Name')['Avg_Recovery_Rate'].mean().reset_index()
avg_recovery_rate_post_2021 = avg_recovery_rate_post_2021.rename(columns={'Avg_Recovery_Rate': 'Avg_Recovery_Rate_Post_2021'})

recovery_rate_change = pd.merge(avg_recovery_rate_pre_2021, avg_recovery_rate_post_2021, on='Disease_Name', how='outer')
recovery_rate_change['Recovery_Rate_Change'] = recovery_rate_change['Avg_Recovery_Rate_Post_2021'] - recovery_rate_change['Avg_Recovery_Rate_Pre_2021']

# Calculate average recovery rate and total death count for each number of comorbidities
avg_recovery_rate_by_comorbidities = df_medicine_sales.groupby('Num_Comorbidities')['Recovery_Rate_by_Disease_Year'].mean().reset_index()
total_death_count_by_comorbidities = df_death_records.groupby('Num_Comorbidities')['Death_Count'].sum().reset_index()

# Calculate hospital type metrics
df_hospital_type_metrics = df_medicine_sales.groupby('Hospital_Type').agg(
    Avg_Recovery_Rate=('Recovery_Rate_by_Disease_Year', 'mean'),
    Total_Revenue=('Total_Revenue', 'sum')
).reset_index()

# Prepare data for regional analysis
regions_to_analyze = df_medicine_sales['Region'].unique()
df_death_records_filtered = df_death_records[df_death_records['State'].isin(regions_to_analyze)].copy()
df_death_records_filtered = df_death_records_filtered.rename(columns={'State': 'Region'})

df_deaths_aggregated_region = df_death_records_filtered.groupby(['Year', 'Disease_Name', 'Region'])['Death_Count'].sum().reset_index()
df_medicine_aggregated_region = df_medicine_sales.groupby(['Year', 'Disease_Name', 'Region']).agg(
    Total_Units_Sold=('Units_Sold', 'sum'),
    Avg_Recovery_Rate=('Recovery_Rate_by_Disease_Year', 'mean'),
    Avg_Govt_Spending_Million=('Govt_Spending_Million', 'mean')
).reset_index()

df_merged_regional_data = pd.merge(df_deaths_aggregated_region, df_medicine_aggregated_region, on=['Year', 'Disease_Name', 'Region'], how='inner')

df_region_metrics = df_merged_regional_data.groupby('Region').agg(
    Avg_Recovery_Rate=('Avg_Recovery_Rate', 'mean'),
    Death_Count=('Death_Count', 'sum'),
    Total_Units_Sold=('Total_Units_Sold', 'sum')
).reset_index()

df_region_metrics_sorted = df_region_metrics.sort_values(by='Avg_Recovery_Rate', ascending=False)

# Identify state with highest death count for each disease
df_deaths_by_disease_state = df_death_records.groupby(['Disease_Name', 'State'])['Death_Count'].sum().reset_index()
highest_death_state_per_disease_actual = df_deaths_by_disease_state.loc[df_deaths_by_disease_state.groupby('Disease_Name')['Death_Count'].idxmax()].reset_index(drop=True)

# Merge for spending vs recovery/sales analysis
df_spending_aggregated = df_medicine_sales.groupby(['Year', 'Disease_Name'])['Govt_Spending_Million'].mean().reset_index()
df_merged_spending = pd.merge(df_merged_data, df_spending_aggregated, on=['Year', 'Disease_Name'], how='left')
```

```python
import numpy as np
from scipy import stats
import pandas as pd

# --- Hypothetical Data ---
# Let's create two hypothetical groups with recovery rates
# Group A: Patients received Treatment A
# Group B: Patients received Treatment B

np.random.seed(42) # for reproducibility
group_a_recovery = np.random.normal(loc=0.75, scale=0.1, size=100) # Mean 75% recovery, 100 patients
group_b_recovery = np.random.normal(loc=0.70, scale=0.1, size=120) # Mean 70% recovery, 120 patients

# Ensure recovery rates are between 0 and 1
group_a_recovery = np.clip(group_a_recovery, 0, 1)
group_b_recovery = np.clip(group_b_recovery, 0, 1)

print("Mean recovery rate for Group A:", np.mean(group_a_recovery))
print("Mean recovery rate for Group B:", np.mean(group_b_recovery))
```

### Performing an Independent Samples t-test

**Research Question:** Is there a significant difference in the average recovery rates between Group A and Group B?

**Hypotheses:**
*   **Null Hypothesis (H0):** There is no significant difference in the average recovery rates between Group A and Group B (Î¼A = Î¼B).
*   **Alternative Hypothesis (H1):** There is a significant difference in the average recovery rates between Group A and Group B (Î¼A â‰  Î¼B).

**Significance Level (alpha):** Let's choose Î± = 0.05.

```python
# Perform independent samples t-test
# We use equal_var=False if we assume unequal variances between the groups (Welch's t-test)
# Otherwise, use equal_var=True
t_statistic, p_value = stats.ttest_ind(group_a_recovery, group_b_recovery, equal_var=False)

print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print(f"\nSince the p-value ({p_value:.3f}) is less than alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the average recovery rates between Group A and Group B.")
else:
    print(f"\nSince the p-value ({p_value:.3f}) is greater than or equal to alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the average recovery rates between Group A and Group B.")
```

### Interpreting the Results

*   **T-statistic:** This value indicates the magnitude of the difference between the group means relative to the variation within the groups.
*   **P-value:** This is the probability of observing a t-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true. A small p-value suggests that the observed difference is unlikely to have occurred by chance.
